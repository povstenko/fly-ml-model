## Importing data

```python
import pandas as pd

data = pd.read_excel("datasets/trips_data.xlsx", index_col=0)
```

```python
data
```

```python
prepared_data = pd.get_dummies(data, columns=['city', 'vacation_preference', 'transport_preference'])
prepared_data
```

```python
data_input = prepared_data.drop('target', axis=1)
```

```python
data_output = prepared_data.target
```

## Create model

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
forest = RandomForestClassifier()
forest.fit(data_input, data_output)
```

```python
forest.predict(data_input)
```

```python
forest.predict_proba(data_input)
```
